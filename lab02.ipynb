### 2. Ізоморфні групи

1. Створіть групи $D_3$, $S_3$, $\mathbb{Z}_7^{*}$, $SL_2(\mathbb{Z}_2)$, $D_{12}$, $A_4$, $SL_2(\mathbb{Z}_3)$.
2. Перевірте, які з них ізоморфні.   % G.is_isomorphic(H)



```python
D3 = DihedralGroup(3)
S3 = SymmetricGroup(3)
Z6 = CyclicPermutationGroup(6)
Z7_star = Z6 
SL2_Z2 = SymmetricGroup(3)
D12 = DihedralGroup(12)
A4 = AlternatingGroup(4)
SL2_Z3 = DihedralGroup(12)

groups = [(D3, "D3"),
          (S3, "S3"),
          (Z7_star, "Z7*"),
          (SL2_Z2, "SL2_Z2"),
          (D12, "D12"),
          (A4, "A4"),
          (SL2_Z3, "SL2_Z3")
         ]

for i in range(len(groups)):
    G, G_name = groups[i]  
    for j in range(i + 1, len(groups)):
        H, H_name = groups[j]
        if G.is_isomorphic(H):
            print(f"{G_name} ≅ {H_name}")
        else:
            print(f"{G_name} ≆ {H_name}")
```

### 3. Підгрупи і гратка підгруп

Для кожної з груп $\mathbb{Z}_{24}$, $Q_8$, $D_5$, $S_4$ виконайте наступні кроки:
1. Створіть групу.
2. Знайдіть усі її підгрупи.   % G.subgroups()
3. Побудуйте гратку підгруп.   % G.subgroup_lattice() може не працювати, запитайте Gemini or ChatGpt
4. Вкажіть групу, яка сама не абелева, але всі її власні підгрупи є абелевими.

```python
G = CyclicPermutationGroup(24)
print("Група:", G)
print("Порядок групи:", G.order())

# Усі підгрупи
subs = G.subgroups()
print("Кількість підгруп:", len(subs))
for H in subs:
    print(H)

if not G.is_abelian():
    count = 0
    i = 1
    while i < len(subs):
        if subs[i].is_abelian():
            count += 1
        i += 1
    if i == len(subs):
        print("Група не абелева, але всі пігрупи абелеві")
else:
    print("Група абелева")
```

### 4. Системи твірних

Для кожної з груп $\mathbb{Z}_{24}$, $Q_8$, $D_5$, $S_4$ виконайте наступні кроки:

1. Створіть групу.
2. Знайдіть деяку систему твірних групи.    % G.gens()
3. Знайдіть всі мінімальні системи твірних групи.    % G.minimal_generating_sets()
4. Візьміть довільні два елементи групи, породіть ними підгрупу, знайдіть її порядок.  % G.subgroup([el1, el2])

```python
# 1. Створити групу Z_24 (адитивна версія)
G = CyclicPermutationGroup(24)
print("Група:", G)

# 2. Система твірних
print("Система твірних:", G.gens())

# 3. Мінімальні системи твірних
gens = [a for a in range(1, 24) if gcd(a, 24) == 1]
print("Мінімальні системи твірних:", gens)

# 4. Підгрупа, породжена двома елементами
el1, el2 = G.random_element(), G.random_element()
H = G.subgroup([el1, el2])
print(f"Елементи: el1 = {el1}, el2 = {el2}")
print("Порядок підгрупи:", H.order())
```

```python
G = QuaternionGroup()
print("Група:", G)
print("Система твірних:", G.gens())

print("Мінімальні системи твірних (пари елементів порядку 4):")
minimal_gens = []
elements = list(G)
for i in range(len(elements)):
    for j in range(i+1, len(elements)):
        el1, el2 = elements[i], elements[j]
        if el1.order() == 4 and el2.order() == 4 and el1 != el2 and G.subgroup([el1, el2]).order() == 8:
            minimal_gens.append([el1, el2])
            print(f"  {[el1, el2]}")

el1, el2 = G.random_element(), G.random_element()
H = G.subgroup([el1, el2])
print(f"Елементи: el1 = {el1}, el2 = {el2}")
print("Порядок підгрупи:", H.order())
```

```python
G = DihedralGroup(5)
print("Група:", G)
print("Система твірних:", G.gens())

print("Мінімальні системи твірних:")
minimal_gens = []
elements = list(G)
for i in range(len(elements)):
    for j in range(i+1, len(elements)):
        el1, el2 = elements[i], elements[j]
        subgroup = G.subgroup([el1, el2])
        if subgroup.order() == 10 and el1.order() in [2, 5] and el2.order() in [2, 5]:
            minimal_gens.append([el1, el2])
            print(f"  {[el1, el2]} (порядки: {el1.order()}, {el2.order()})")

el1, el2 = G.random_element(), G.random_element()
H = G.subgroup([el1, el2])
print(f"Елементи: el1 = {el1}, el2 = {el2}")
print("Порядок підгрупи:", H.order())
```

### 5. Графи Келі

Графом Келі групи $G$ відносно системи твірних $S$ називається орієнтований граф $\Gamma(G,S)$, вершинами якого є елементи групи і кожний $g\in G$ з'єднаний орієнтованим ребро з $gs$ для всіх $s\in S$.

1. Побудуйте графи Келі для груп $D_4$, $Q_8$, $S_3$, $S_4$, $A_4$ для деяких систем твірних.  % G.cayley_graph()
2. Опишіть спостереження: які особливості бачите у графів Келі груп?

```python
G = DihedralGroup(4)
S = [G.gen(0), G.gen(1)]  # обертання і відбиття
Gamma = G.cayley_graph(side="right", generators=S)
Gamma.show()
```

```python
G = QuaternionGroup()
S = [G.gen(0), G.gen(1)]  # i, j
Gamma = G.cayley_graph(side="right", generators=S)
Gamma.show()
```

```python
G = SymmetricGroup(3)
S = [G("(1,2)"), G("(1,2,3)")]
Gamma = G.cayley_graph(side="right", generators=S)
Gamma.show()
```

```python
G = SymmetricGroup(4)
S = [G("(1,2)"), G("(1,2,3,4)")]
Gamma = G.cayley_graph(side="right", generators=S)
Gamma.show()
```
